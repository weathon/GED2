```python
with open("out","r") as f:
    data = f.read().split("\n")

if "OOBA" in "".join(data):
    print("FUCK YOU!")
    quit()

vTreeTime = []
eTreeTime = []
BFTime = []
BaselineTime= [] 
for i in data:
    if i.split(":")[0] == "VTree time":
        vTreeTime.append(float(i.split(":")[1]))
    if i.split(":")[0] == "ETree time":
        eTreeTime.append(float(i.split(":")[1]))
    if i.split(":")[0] == "BF time":
        BFTime.append(float(i.split(":")[1]))
    if i.split(":")[0] == "Baseline time":
        BaselineTime.append(float(i.split(":")[1]))
        

import pylab
pylab.scatter(range(len(vTreeTime)),vTreeTime, label="VTree")
pylab.scatter(range(len(vTreeTime)),eTreeTime, label="ETree")
pylab.scatter(range(len(vTreeTime)),BFTime, label="BF")
pylab.scatter(range(len(vTreeTime)),BaselineTime, label="Baseline")
pylab.legend()
```
